# 1. Load _contsep_ class
## It will download PS reference frame.

```python
date = "20200127"
targetid = "ZTF20aagzchj"
```

```python
test_ = contsep.get_spaxels_from_constsep(date, targetid)
```

# 2. Set IFU data "offset" from the reference PS data.
### `offset` should be given by hand after visual inspection.

### !! Will be automated. !!

```python
# default is offset=(0.0, 0.0)
test_.show_ref_ifu( offset=(0.0, 0.0))
```

```python
# For this target, offset=(2.3, -0.3)
test_.show_ref_ifu( offset=(2.3, -0.3))
```

# 3. Get target OR others (includin host) spaxels

## -For the target 'spaxel ids',

```python
target_spaxel_ids = test_.get_target_spaxels(spaxels_id=True) # spaxels_id=True is default.
target_spaxel_ids
```

### You can also get target 'spaxel indexes' with _`spaxels_id=False`_.
### Spaxel indexes are used in `pysedm`.

```python
target_spaxel_indexes = test_.get_target_spaxels(spaxels_id=False)
target_spaxel_indexes
```

## -For others (including host)

```python
host_spaxel_ids = test_.get_others_spaxels(spaxels_id=True)
host_spaxel_ids
```

### 'spaxel indexes' for other source,

```python
host_spaxel_indexes = test_.get_others_spaxels(spaxels_id=False)
host_spaxel_indexes
```
